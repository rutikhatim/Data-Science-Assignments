# Prepare Rules for all data sets

 ### For book dataset


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
book=pd.read_csv("book.csv")
book
```

```python
# data preprocessing isn't required as it is already in the transaction format
```

# Apriori Algorithm

# Association rule with 10% support and 70% confidence

```python
# With 10% Support
frequent_itemsets=apriori(book,min_support=0.1,use_colnames=True)
frequent_itemsets
```

```python
# with 70% confidence
rules=association_rules(frequent_itemsets,metric='lift',min_threshold=0.7)
rules
```

A leverage value of 0 indicates independence.

Range will be A high conviction value means that the consequent is highly depending on the antecedent and range [0 inf]

```python
rules.sort_values('lift',ascending=False)
```

```python
# Lift Ratio > 1 is a good influential rule in selecting the associated transactions
rules[rules.lift>1]
```

```python
# visualization of obtained rule
plt.scatter(rules['support'],rules['confidence'])
plt.xlabel('support')
plt.ylabel('confidence') 
plt.show()
```
